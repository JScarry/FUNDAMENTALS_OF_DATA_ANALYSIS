Write a Python function to remove all non-alphanumeric characters from a string.

### Exercise 1 answer:
Defined the function fn that uses the Python Regex sub() function to find and replace. 
The syntax is regex.sub(pattern,replacement,output string) 

Ref# https://www.pythontutorial.net/python-regex/python-regex-sub/

```python
import re
s = 'The function "fn" should remove all non-alphanumeric characters from this string, 321.. here goes, fingers crossed! ;-)'


def fn():
    s2 = re.sub(r'[^a-z\d]',' ', s, flags=re.IGNORECASE) #re.sub substitutes the pattern with the replacement in the string.
                                        #in this case [^a-z+A-Z+\d] = ( ^ = everything except chars a-z and decimal 0-9 )
                                        #will be substituted with a blank space. flags=re.IGNORECASE will include upper and
                                        #lower case a-z.
                                        #note: this will find and replace blank spaces also!
                                        
    print('After:',s2)  #fn output


print('Before:' ,s) #print original string before calling fn
fn() 
```

Find the first letter in the word:

Called the function in the re.sub repl argument, this replaces ([a-z]+) with output of the function.

Tried to run function in re.sub. Removed the complile in an attempt to simplify it. Had trouble trying to use a variable name in re.sub. #https://stackoverflow.com/questions/56600312/use-a-variable-name-in-re-sub

```python
# https://stackoverflow.com/a/1393367

import urllib.request

url = r'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

iris = [line.decode('utf-8').strip() for line in urllib.request.urlopen(url)]

iris
```

```python
strip_iris = re.compile(r'([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),([0-9]\.[0-9]),Iris-([a-zA-Z]+)')
cap_iris = [strip_iris.sub(r'\5,\4,\3,\2,\1', line) for line in iris if line]
def f(find):
    return find.group().capitalize()
[re.sub(r'\b([a-z]+)', f, line) for line in cap_iris]
```
