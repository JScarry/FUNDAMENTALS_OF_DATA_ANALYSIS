A Normal Distribution graph or Bell curve as it is commonly known. 
Common use of the bell curve is in test score results.


The area under the graph represents 100% of the data
Empirical rule. Data falls within a certain number of standard deviations from the mean [1] [2]
68% =1 (sigma)
95% = 2sd
99.7% = 3sd

There are a few basic properties of a normal distribution.
The graph will be Symmetrical
The Mean, Mode and Median are all equal
