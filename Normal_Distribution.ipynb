## WHAT IS NORMAL DISTRIBUTION

Normal distribution is a probability distribution that is symmetric about the mean, 
showing that data near the mean are more frequent in occurrence than data far from the mean. 
It is also known as the Gaussian distribution after the German mathematician Carl Friedrich Gauss who first described the phenomen.

In graphical form, the normal distribution appears as a "bell curve".  
An example is shown below. [13]

A good summary then is; The term Normal Distribution Curve or Bell Curve is used to describe the mathematical 
concept called normal distribution, sometimes referred to as Gaussian distribution. 
It refers to the shape that is created when a line is plotted using the data points 
for an item that meets the criteria of Normal Distribution.

The general form of its probability density function is shown below [3]

## Properties of a Normal distribution bell curve

The area under the graph represents 100% of the data
Empirical rule. Data falls within a certain number of standard deviations from the mean [1] [2]
68% =1 (sigma)
95% = 2
99.7% = 3

There are a few basic properties of a normal distribution.

The graph will be Symmetrical.

The Mean, Mode and Median are all equal.

The standard normal distribution has a mean of 0 and a variance of 1 standard deviation

## The Empirical rule of standard deviation. 

In a standard normal distribution:
68% of the data will fall within one standard deviation from the mean
95% of the data will fall within two standard deviations
99.73% of the data points will fall within three standard deviations from the mean
99.9937% of the data points will fall within four standard deviations from the mean
99.999943% of the data points will fall within five standard deviations from the mean
99.9999998% of the data points will fall within six standard (6 Sigma) deviations from the mean

## 6   

Normal distribution is used in industary whare a high degree of accuracy is required, such as in medical device manufacture. The term 6 sigma or 6 is commonly used and it refers to a quality control philosophy which has its roots found in normal distribution. Put simply, the aim during manufacture is that the process should operate within 6 sigma at all times. Measurments are taken and during the process and results should show that 99.9999998% of items fall within the specifications. 

## Central limit theorem

## Normal distribution examples using Python Numpy

The code below uses Numpy to generate random numbers in a normal distribution [11]

3 parameters in the function. These allow us to control the mean, the standard deviation, and the size of the normal distribution

A random sample of 10 numbers, mean = 0 and standard deviation of 0.1 is printed as an example, then random_data is generated but this time with 1000 samples.

The output chart shows a distribution plot and a histogram of the generated random_data. 

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

random_data_sample = random.normal(0, 0.1, 10)

random_data = random.normal(0, 0.1, 1000)
sns.distplot(random_data, hist=True, bins=100)
print(random_data_sample)
plt.show()
```

## Normal distribution examples using Python Numpy

The code below [10] uses Numpy to generate random numbers in a normal distribution. 

We see that the larger the sample size (100,000 in this case) the more defined the shape of the "Bell curve". The distribution is more "normal"

```python
import numpy as np

mu, sigma = 0, .1 # mean and standard deviation
s = np.random.default_rng() .normal(mu, sigma, 100000)
#print(s)

import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 100, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ))
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

## Is my data normal?

We can run a check on a set of data to see if it fits the "normal distribution. In this case I took the familiar iris dataset and ran the setosa petal length values through the "Fitter" library

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import time
from fitter import Fitter

iris_df = sns.load_dataset("iris") #load iris dataset directly from seaborn
print(iris_df.head())
print()
print(iris_df.info()) #display df info
print()
setosa_df = iris_df.iloc[1:50] #select first 50 rows (all setosa entries)
versicolor_df = iris_df.iloc[51:101]
virginica_df = iris_df.iloc[101:151]
sns.set_style('white')
sns.set_context("paper", font_scale = 2) #setup style
sns.displot(data=setosa_df, x="sepal_length", kind="hist", bins = 50, aspect = 1.5,) #plot histogram of setosa sepal lengths 
sns.displot(data=versicolor_df, x="sepal_length", kind="hist", bins = 50, aspect = 1.5,)
sns.displot(data=virginica_df, x="sepal_length", kind="hist", bins = 50, aspect = 1.5,)

sepal_length_fn = setosa_df["sepal_length"].values #create function for sepal length values
#f = Fitter(sepal_length_fn) #check for best fit of data (https://medium.com/the-researchers-guide/finding-the-best-distribution-that-fits-your-data-using-pythons-fitter-library-319a5a0972e9)
f = Fitter(sepal_length_fn,distributions=['gamma','lognorm',"beta","burr","norm"])
f.fit()
f.summary() #print a summmary
```
