## WHAT IS NORMAL DISTRIBUTION

First we could ask, what is statistical distribution? A statistical distribution or probability distribution describes how values are distributed for a field. In other words, the statistical distribution shows which values are common and uncommon. We use a statistical distribution to determine how likely a particular value is. 

[15] (IBM Cognos Analytics, Nov 2022)

"The normal distribution is the single most important distribution in the social sciences" 

[16] (Normal Distribution, encyclopedia.com, May 2018)


It is important because many psychological and educational variables are distributed normally (approximately). For example if we measure job satisfaction, the reading ability of a population or memory we find they are normally distributed. Infact many natural, physical and psychological variables are approximately normally distributed.  Although these distributions are only approximately normal, they are usually very close. A second reason the normal distribution is so important is that it is easy for mathematical statisticians to work with. This means that many kinds of statistical tests can be derived for normal distributions.

### Normal distribution history

It was first observed by Galileo Galilei (1564–1642) when he noticed errors in astronomical observations were not totally random, but instead they had some patterns. In the first decade of the nineteenth century the mathematicians Adrien-Marie Legendre (1752–1833) and the German mathematician Carl Friedrich Gauss (1777–1855) worked out the precise mathematical formula, and Gauss demonstrated that this curve provided a close fit to the empirical distribution of observational errors. 

While the normal distribution was first applied to the description of measurement errors, scientists later began to realize that it also described variation in human phenomena independent of errors of measurement. In 1835 Adolphe Quetelet (1796–1874) applied the normal distribution to many physical attributes, such as height. [16] (Normal Distribution, encyclopedia.com, May 2018)

But it was in 1869 Francis Galton (1822–1911) found the normal distribution could be extended to cover individual differences in ability. For this reason intellegence tests score a persons test based on a normal distribution. It is important to remember that while many, not all human attributes or behavioral events are normally distributed, for example data about alcohol or coffee consumption are usually not bell shaped. These typically follow a right-skewed distribution. They will often have a clusters of values (nonconsumers and consumers and heavy consumers). We need to be aware of whether variables follow normal or non-normal distributions as this will influences how we go about analysing and describing the data.[17] (Dealing With Non-normal Data,  Kristin L. Sainani, Dec 2012)

### Defining normal distribution
Normal distribution is a probability distribution that is symmetric about the mean, showing that data near the mean are more frequent in occurrence than data far from the mean. 

In graphical form, the normal distribution appears as a "bell curve".  
An example is shown below.

$$ p(x) = \frac{1}{\sqrt{2 \pi \sigma^2 }} e^{-\frac{(x - \mu)^2}{2 \sigma^2}} $$

### The Empirical rule of standard deviation. 

In a standard normal distribution:
68% of the data will fall within one standard deviation from the mean
95% of the data will fall within two standard deviations
99.73% of the data points will fall within three standard deviations from the mean
99.9937% of the data points will fall within four standard deviations from the mean
99.999943% of the data points will fall within five standard deviations from the mean
99.9999998% of the data points will fall within six standard (6 Sigma) deviations from the mean

### 6 σ  

Normal distribution is used in industary whare a high degree of accuracy is required, such as in medical device manufacture. The term 6 sigma or 6σ is commonly used and it refers to a quality control philosophy which has its roots found in normal distribution. Put simply, the aim during manufacture is that the process should operate within 6 sigma at all times. Measurments are taken and during the process and results should show that 99.9999998% of items fall within the specifications.

### Probability density function (PDF)

In probability theory, a probability density function (PDF), or density of a continuous random variable, is a function whose value at any given sample (or point) in the sample space (the set of possible values taken by the random variable) can be interpreted as providing a relative likelihood that the value of the random variable would be equal to that sample. 

So the PDF is used to find the probability of the random variable falling within a range of values, rather than being a specific value. 

[19] (Probability Density Function, wikipedia, Jan 2023)

Lets consider the probability of a womans height being exactly 1.7m?

Well if we want to find somone exactly 1.700000m it would be very unlikley, infact we could say that probability is close to Zero. 

However, what is the probability of finding a woman around 1.7m, say we were willing to accept a height between 1.68 to 1.72m? Then we can calculate the probability based on the area under that part of the bell curve.

## Normal distribution examples using Python Numpy

The code below uses Numpy to generate random numbers in a normal distribution. 

[10] (Normal (Gaussian) Distribution, Dec 2022)  

[12] (A Quick Introduction to Numpy Random Normal, Joshua Ebner, June 2022)

3 parameters in the function. These allow us to control the mean, the standard deviation, and the size of the normal distribution

I ran a random sample of 100 numbers, mean = 0 and standard deviation of 0.1 and plotted the random results. The data appears to be normally distributed but the bell curve is not well defined due to the small sample size. When the sample size is increased to 10000 we can see a well defined bell curve 



```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

random_data_100 = random.normal(0, 0.1, 100)

random_data_10000 = random.normal(0, 0.1, 10000)
sns.distplot(random_data_100, hist=True, bins=10).set(title='random_data_100')
plt.show()
sns.distplot(random_data_10000, hist=True, bins=100).set(title='random_data_1000')
#print(random_data_sample)
plt.show()
```

### Sample size
 
 We see again, that the larger the sample size (100,000 in this case) the more defined the shape of the "Bell curve". The distribution is more "normal"

```python
import numpy as np

mu, sigma = 0, .1 # mean and standard deviation
s = np.random.default_rng() .normal(mu, sigma, 100000)
#print(s)

import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 100, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ))
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

### Normal distribution for different $\mu$ and $\sigma$

A good explanation of the effect of $\mu$ and $\sigma$ came up in one of our Programming_for_Data_Analysis lectures with Brian McGinley 

[11] (Code from Programming_for_Data_Analysis lectures, Brian McGinley, Nov 2022) 

The code below plots 4 normal distributions with different values for $\mu$ and $\sigma$. 

In the yellow plot for example, a value from our data is very likley to be close to -1 (which is $\mu$, our mean), then in the green plot a value from our data is less likley to be close to our mean, which in this case is 0. Any value picked from our data is likley to fall within 3 times $\mu$, as seen in the green plot where most values are between -9 and +9.

Also, what is maybe more noticable on the green plot here, but is true for all distributions, the X value does drop towards zero as you move away from the mean, but it does not reach zero but continues to infinity. So in the green plot it may be possible to find a value of -50, but this would be very unlikley.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as ss

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = 14, 10
```

```python
x = np.linspace(-10.0, 10.0, 1000)

plt.fill(x, ss.norm.pdf(x, loc= 0.0, scale=1.0), label="$\mu = 0.0, \sigma = 1.0$", c='b', alpha=0.6, lw=3.0)
plt.fill(x, ss.norm.pdf(x, loc= 2.0, scale=1.0), label="$\mu = 2.0, \sigma = 1.0$", c='r', alpha=0.6, lw=3.0)
plt.fill(x, ss.norm.pdf(x, loc= 0.0, scale=3.0), label="$\mu = 0.0, \sigma = 3.0$", c='g', alpha=0.6, lw=3.0)
plt.fill(x, ss.norm.pdf(x, loc= -1.0, scale=0.5), label="$\mu = -1.0, \sigma = .05$", c='y', alpha=0.6, lw=3.0)

plt.title("Normal distribution for different $\mu$ and $\sigma$")
plt.legend();
```

## Is my data normal?

We can check our data to see if it fits the "normal distribution". In this case I took the familiar iris dataset and ran the setosa sepal length values through the "Fitter" library. We can run the best_fit library which checks our data against a library of over 100 different distributions. This takes about 30 mins to run so I changes it to specified 5 common distributions, including normal to see how they compare. In this case, normal distribution returned the second lowest summsquare error after Burr distribution, so yes I would consider normal distribution to be a good description for setosa sepal length 

[14] (Finding the Best Distribution that Fits Your Data using Python’s Fitter Library, Rahul Raoniar, Jun 2021)

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import time
from fitter import Fitter

iris_df = sns.load_dataset("iris") #load iris dataset directly from seaborn
print(iris_df.tail()) #display the new iris_df tail 

setosa_df = iris_df.loc[iris_df.loc[:,'species'] == 'setosa'] #select only setosa entries from the species column
print()
print(setosa_df.tail()) #display the new setosa_df tail to check it has been selected correctly - only setosa
print()

sns.set_style('white') #set the plot style
sns.set_context("paper", font_scale = 2) #set the plot style
sns.displot(data=setosa_df, x="sepal_length", kind="hist", bins = 50, aspect = 1.5,) #plot histogram of setosa sepal lengths 

sepal_length_fn = setosa_df["sepal_length"].values #create function for sepal length values
f = Fitter(sepal_length_fn,distributions =['gamma','lognorm',"beta","burr","norm"]) #check for best fit of data [14]
#f.get_best(method = 'sumsquare_error') #removed this as it takes 30mins to run
f.fit()
f.summary() #print a summmary
```

##### Observations: 
In this case, normal distribution returned the second lowest summsquare error after Burr distribution, so yes I would consider normal distribution to be a good description for setosa sepal length. With a larger sample size I would expect this natural variable to fall within a normal distribution.

## Other common types of distributions

![image info](https://i.postimg.cc/sDRksm7T/2.png)
[21] (All distribution in statistics, AHMED M MAHROUS, Jan 2022) 


##### Uniform distribution. Example: Rolling a dice. 

Equal chance of any number from 1-6, so with enough samples the distribution is likley to be uniform. 

##### Binomial distribution. Example: Tossing a coin. 

The probability of getting Heads remains the same for the trials we are investigating 

##### Poisson distribution.  Example: Footfall. People to walk into a shop in 1 hour. 

A discrete probability distribution. The probability of something happening a certain number of times within a given interval. 

##### Exponential Distribution. Example: The time between earthquakes. 

It is often concerned with the amount of time until some event occurs. If an earthquake occurs this year, what is the probability of another occurring again next year? 

##### Log-normal distribution. Example: Milk produced per Cow. 

It is a continuous distribution of random variable whose natural logarithm is normally distributed.

##### The T distribution (also called Student’s T Distribution). Estimating average height based on a small sample. 
It is a family of distributions that look almost identical to the normal distribution curve, only a bit shorter and fatter. The t distribution is used instead of the normal distribution when you have small samples and the standard deviation of the population is unknown.

##### Skew normal distribution. Example: Income and wealth are typically right skewed distributions. 
Many people earn a small to medium amount, but some wealthy and a few extremley wealthy extend the right tail into very high values. Meanwhile, the left tail cannot be less than zero. This situation creates a right skew, or positive skew in the distribution.
